```python
def dfs(G, start , visited = None):  
  print (start)
  if visited is None:
    visited = set() 
  visited.add(start)
  for next in G[start] - visited:
      dfs(G, next, visited)
  return visited

G = {'0': set(['1','2','3']),
        '1': set(['0','2']),
        '2': set(['0','1','4']),
        '3': set(['0']),
        '4': set(['2'])
        }

dfs(G, '0')
```

```python
visited = set() 
def dfs(visited, g, node): 
    if node not in visited:
        print (node)
        visited.add(node)
        for next in g[node]:
            dfs(visited, g, next)

g = {
        '0': set(['1','2','3']),
        '1': set(['0','2']),
        '2': set(['0','1','4']),
        '3': set(['0']),
        '4': set(['2'])
        }            

dfs(visited, g, '0')
```

```python
def printtabel(tabel):
  for i in tabel:
    for j in i:
      print(str(j)+" ", end=" ")
    print()

str1 = ["A", "B", "D", "F", "R", "S", "T"]
str2 = ["A", "B", "C", "F", "O", "R", "T"]

len_str1 = len(str1)
len_str2 = len(str2)
               
print(len_str1,len_str2)
print("============")

tabel = [[0 for x in range(len_str1+1)] for y in range(len_str2+1)]
for i in range(len_str1):
  for j in range(len_str2):
    if str2[j] == str1[i]:
      tabel[i+1][j+1] = tabel[i][j]+1
    else:
      if tabel[i][j+1] > tabel[i+1][j]:
        tabel[i+1][j+1] =  tabel[i][j+1]
      else:
        tabel[i+1][j+1] = tabel[i+1][j]

printtabel(tabel)
print("============")

i= len_str1
j = len_str2
while(1):
  #print(tabel[i][j])
  if(tabel[i-1][j] == tabel[i][j-1]):
    if tabel[i][j] > tabel[i-1][j-1]:
      print(str1[i-1]+" ", end = "" )
    i-=1
    j-=1
  elif(tabel[i-1][j] > tabel[i][j-1]):
    i-=1
  else:
    j-=1
  if i<0 or j<0:
    break
```
