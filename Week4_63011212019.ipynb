```python
import numpy as np
import random

def eudist(x,y): 
  return np.sqrt((x-y)**2)
  
x = [7,25,35,75,74,65,34,23,95,11,14,23,52,67,41,33,14] 
k = 2

k1 = 20
k2 = 40

p = 0
while(1):
  p+=1
  l1 = [ ]
  l2 = [ ]
  for v in x:
    dist1 = eudist(k1,v)
    dist2 = eudist(k2,v)
    if(dist1 <= dist2):#ถ้า dist1 น้อยกว่า dist2
      l1.append(v)#เพิ่มข้อมูลลงในตัวแปร List l1
    else:
      l2.append(v)#เพิ่มข้อมูลลงในตัวแปร List l2
  m1 = np.mean(l1)#หาค่าเฉลี่ย l1
  m2 = np.mean(l2)#หาค่าเฉลี่ย l2

  if(m1 == k1 and m2 == k2): 
    break
  else:
    k1 = m1
    k2 = m2
    print("K1 = ",k1)
    print("K2 = ",k2)
    print("Round: "+str(p))
    print(l1)
    print(l2)
    print("-----------------------------------------------------")
```

```python
import numpy as np
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt

def dist(x,y):
  return np.sqrt(np.sum((np.array(x) - np.array(y)) ** 2))

data = make_blobs(n_samples = 1000, n_features = 2, centers = 4, cluster_std = 3, random_state = 100)

k = 2
k1 = [-1, 25]
k2 = [-7, 1]
p = 0
e = 0.02

while(1):
  p += 1
  l1 = [ ]
  l2 = [ ]
  for v in data[0]:
    d1 = dist(k1, v)
    d2 = dist(k2, v)
    if(d1 <= d2):
      l1.append(v)
    else:
      l2.append(v)
  m1 = np.mean(np.array(l1), axis = 0)
  m2 = np.mean(np.array(l2), axis = 0)
  if dist(m1, k1) < e and dist(m2, k2) < e :
    break
  else:
    k1 = m1
    k2 = m2
    print("K1 = ",k1)
    print("K2 = ",k2)
    print("Round : "+str(p))
    print(l1)
    print(l2)
    plt.scatter(np.array(l1)[:,0],np.array(l1)[:,1], marker = "*", s = 10, c = 'red')
    plt.scatter(np.array(l2)[:,0],np.array(l2)[:,1], marker = "*", s = 10, c = 'blue')
    plt.show()
```

```python
import cv2 
from google.colab.patches import cv2_imshow
from sklearn.cluster import KMeans

img = cv2.imread('/content/drive/MyDrive/AI/Pic_AI.jpg',0)

row, col  = img.shape
dat = img.reshape(1, row*col)
Kmean = KMeans(n_clusters = 4)
output = Kmean.fit(dat[0].reshape(-1,1))
coler = [[227, 63, 28], [61, 227, 28], [227, 224, 28],[137, 28, 227],[28, 212, 227],[255, 0, 0 ]]
output_img = np.zeros((row,col,3), dtype = np.uint8)
k = 0
for i in range (row):
  for j in range(col):
    output_img[i,j] = coler[int(output.labels_[k])]
    k += 1
cv2_imshow(output_img)
```
