```python
import random
b=[]
while 1:
  flag =1
  n = random.randint(0,9)
  for x in b:
    if n==x:
      flag = 0
      break
  if flag==1:
    b.append(n)
  if len (b)==9:
    break
print(b)
```

```python
def showState(b):
  for i,x in enumerate(b):
    if i%3 == 0 and i != 0:
        print()
    print(str(x)+"   ",end=" ")
  print()
  print()

def moveleft (b):
  z = b.index(0)
  ml = [0,3,6]
  if z in ml:
    return b
  x = b[z]
  b[z] = b[z-1]
  b[z-1] = x
  return b

def moveright (b):
  z = b.index(0)
  mr = [2,5,8]
  if z in mr:
    return b
  x = b[z]
  b[z] = b[z+1]
  b[z+1] = x
  return b

def moveup (b):
  z = b.index(0)
  mu = [0,1,2]
  if z in mu:
    return b
  x = b[z]
  b[z] = b[z-3]
  b[z-3] = x
  return b

def movedown (b):
  z = b.index(0)
  md = [6,7,8]
  if z in md:
    return b
  x = b[z]
  b[z] = b[z+3]
  b[z+3] = x
  return b

b = [1,2,3,8,0,4,7,6,5]
showState(b)

moveleft(b)
showState(b)

moveright(b)
showState(b)

moveup(b)
showState(b)

movedown(b)
showState(b)
```

```python
class Node:
  def __init__(self) :
      self.parent: Node = self
      self.childrean: List[Node] = []
      self.data = []
  
  def add_child(self):
      child = Node()
      self.childrean.append(child)
      child.parent =  self
      return child

b = [1,2,3,8,0,4,7,6,5]
tree = Node()
tree.data = b
print("A")
showState(tree.data)

print("A.1 left")
tree.add_child()
tree.childrean[0].data = moveleft(b.copy())
showState(tree.childrean[0].data)

print("A.2 right")
tree.add_child()
tree.childrean[1].data = moveright(b.copy())
showState(tree.childrean[1].data)

print("A.3 up")
tree.add_child()
tree.childrean[2].data = moveup(b.copy())
showState(tree.childrean[2].data)

print("A.4 down")
tree.add_child()
tree.childrean[3].data = movedown(b.copy())
showState(tree.childrean[3].data)

print("A.1.1 up")
tree.childrean[0].add_child()
tree.childrean[0].childrean[0].data = moveup(tree.childrean[0].childrean[0].parent.data)
showState(tree.childrean[0].childrean[0].data)

print("A.1.2 down")
tree.childrean[0].add_child()
tree.childrean[0].childrean[0].data = movedown(tree.childrean[0].childrean[0].parent.data)
showState(tree.childrean[0].childrean[0].data)

print("A.2.1 moveup")
tree.childrean[1].add_child()
tree.childrean[0].childrean[0].data =moveup(tree.childrean[0].childrean[0].parent.data)
showState(tree.childrean[0].childrean[0].data)

print("A.2.2 movedown")
tree.childrean[1].add_child()
tree.childrean[0].childrean[0].data = movedown(tree.childrean[0].childrean[0].parent.data)
showState(tree.childrean[0].childrean[0].data)

print("A.3.1 up")
tree.childrean[2].add_child()
tree.childrean[0].childrean[0].data = moveup(tree.childrean[0].childrean[0].parent.data)
showState(tree.childrean[0].childrean[0].data)

print("A.3.2 down")
tree.childrean[2].add_child()
tree.childrean[0].childrean[0].data = movedown(tree.childrean[0].childrean[0].parent.data)
showState(tree.childrean[0].childrean[0].data)

print("A.4.1 right")
tree.childrean[3].add_child()
tree.childrean[0].childrean[0].data = moveright(tree.childrean[0].childrean[0].parent.data)
showState(tree.childrean[0].childrean[0].data)

print("A.4.2 left")
tree.childrean[3].add_child()
tree.childrean[0].childrean[0].data = moveleft(tree.childrean[0].childrean[0].parent.data)
showState(tree.childrean[0].childrean[0].data)
```
