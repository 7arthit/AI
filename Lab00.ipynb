```python
import re
value = [ ]
items = [x for x in input().split(',')]
for p in items:
  if len(p)<6 or len(p)>12:
      continue
  else:
      pass
  if not re.search("[a-z]",p):
       continue
  elif not re.search("[0-9]",p):
      continue
  elif not re.search("[A-Z]",p):
       continue
  elif not re.search("[$#@]",p):
      continue
  elif re.search("\s",p):
      continue
  else:
      pass
  value.append(p)
print (",".join(value))
```

```python
import math
pos = [0,0]
while True:
    s = input()
    if not s:
        break
    movement = s.split(" ")
    direction = movement[0]
    steps = int(movement[1])
    if direction == "UP":
        pos[0] += steps
    elif direction == "DOWN":
        pos[0] -= steps
    elif direction == "LEFT":
        pos[1] -= steps
    elif direction == "RIGHT":
        pos[1 ]+= steps
    else:
        pass

print (int(round(math.sqrt(pos[1]**2+pos[0]**2))))
```
